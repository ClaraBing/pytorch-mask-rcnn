```python
from maskRCNN.pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
annot_path = '/vision/u/cy3/data/EPIC/annotations/coco_train_object_labels_819.json'
#annot_path = '/vision/u/bingbin/EPIC_KITCHENS_2018/annotations/coco_train_object_labels.json'
coco = COCO(annot_path)
```

```python
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))
print(len(cats))
```

```python
catIds = coco.getCatIds(catNms=['fridge', 'pan'])
imgIds = coco.getImgIds(catIds=catIds )
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
```

```python
plt.imshow(I); plt.axis('off')
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
```
